```python
def removeSpace(file):
    new_names=[]
    for col_name in file.columns:
        col_name=str.lower(col_name)
        col_name = str.strip(col_name)
        str.replace(col_name,'\s','_')
        new_names[len(new_names):]=[col_name]
        
        
    file.columns = new_names
```
